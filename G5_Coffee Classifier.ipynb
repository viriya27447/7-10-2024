### 1.1 Problem and Pain point
ในการแยกเมล็ดกาแฟคั่วด้วยวิธีการแบบดั้งเดิม มักจะอาศัยแรงงานคนในการคัดเลือกเมล็ดคั่วระดับที่ต้องการ ซึ่งกระบวนการนี้มีข้อจำกัดและปัญหาหลายประการ

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
#---------------------------------------
import tensorflow as tf
from tensorflow.keras.layers import Flatten, Dense, Dropout
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.optimizers import Adamax
from tensorflow.keras.metrics import Precision, Recall
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping
#---------------------------------------
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def train_df(tr_path):
    classes = []
    class_paths = []
    files = os.listdir(tr_path)
    for file in files:
        label_dir = os.path.join(tr_path, file)
        label = os.listdir(label_dir)
        for image in label:
            image_path = os.path.join(label_dir, image)
            class_paths.append(image_path)
            classes.append(file)
    image_classes = pd.Series(classes, name='Class')
    image_paths = pd.Series(class_paths, name='Class Path')
    tr_df = pd.concat([image_paths, image_classes], axis=1)
    return tr_df
```

```python
def test_df(ts_path):
    classes = []
    class_paths = []
    files = os.listdir(ts_path)
    for file in files:
        label_dir = os.path.join(ts_path, file)
        label = os.listdir(label_dir)
        for image in label:
            image_path = os.path.join(label_dir, image)
            class_paths.append(image_path)
            classes.append(file)
    image_classes = pd.Series(classes, name='Class')
    image_paths = pd.Series(class_paths, name='Class Path')
    ts_df = pd.concat([image_paths, image_classes], axis=1)
    return ts_df
```

**Input train&test data**

```python
tr_df = train_df(r'D:\coffee\train')
```

```python
ts_df = test_df(r'D:\coffee\test')
```

```python
batch_size = 16
img_size = (224,224)

tr_gen = ImageDataGenerator(
    brightness_range=(.8, 1.2)  # ความสว่างของภาพจะถูกปรับในช่วง 80% ถึง 120%
)
```

**<a id="section-233">Data Generator</a>**

 **Training**

```python
tr_gen = tr_gen.flow_from_dataframe(
    dataframe=tr_df,
    x_col='Class Path',
    y_col='Class',
    batch_size=batch_size,
    target_size=img_size,
    class_mode='categorical'
)
```

**Validation**

```python
# สร้าง Data Generator สำหรับ validation set จาก DataFrame
valid_gen = ImageDataGenerator().flow_from_dataframe(
    dataframe=valid_df,         # DataFrame ของ validation set
    x_col='Class Path',
    y_col='Class',
    batch_size=batch_size,
    target_size=img_size,
    class_mode='categorical'
)
```

**Test**

```python
# สร้าง Data Generator สำหรับ test set จาก DataFrame
ts_gen = ImageDataGenerator().flow_from_dataframe(
    dataframe=ts_df,            # DataFrame ของ test set
    x_col='Class Path',
    y_col='Class',
    batch_size=batch_size,
    target_size=img_size,
    class_mode='categorical',
    shuffle=False               # ไม่ทำการ shuffle ข้อมูลในชุด test
)
```

```python
class_dict = tr_gen.class_indices
classes = list(class_dict.keys())
images, labels = next(tr_gen)

plt.figure(figsize= (20, 20))

for i in range(16):
    plt.subplot(4,4,i+1)
    image = images[i] / 255
    plt.imshow(image)
    index = np.argmax(labels[i])
    class_name = classes[index]
    plt.title(class_name, color='k', fontsize=20)
plt.show()
```

learning_rate = 0.001 ค่านี้เหมาะกับการเริ่มฝึกโมเดล เช่น Adam ซึ่งจะทำให้โมเดลเรียนรู้แบบค่อยเป็นค่อยไปและลด oscillation ระหว่างการฝึก
ลดปัญหา Overfitting ด้วย dropout_rate_1 = 0.3, dropout_rate_2 = 0.25 หมายถึง อัตรากาารสุ่มปิดหน่วยประสาท 30% และ 25% แต่ละชั้น
dense_units = 128: จำนวนหน่วยประสาทในชั้น Dense Layer ถูกตั้งค่าไว้ที่ 128 หน่วย ซึ่งเป็นค่าที่ค่อนข้างมาตรฐานสำหรับการเริ่มต้น(หากปัญหาซับซ้อนก็สามารถปรับได้)
epochs = 10, batch_size = 32: การฝึกโมเดลจะทำซ้ำ 10 Epoch โดยแบ่งข้อมูลออกเป็นชุดย่อย (batch) ขนาด 32 ตัวอย่างในการคำนวณแต่ละครั้ง

```python
img_shape = (224, 224, 3)
num_classes = len(classes)

# Hyperparameters สำหรับโมเดล
learning_rate = 0.001
dropout_rate_1 = 0.3
dropout_rate_2 = 0.25
dense_units = 128

# จำนวน epochs และ batch size สำหรับการฝึกโมเดล
epochs = 10
batch_size = 32
```

### 3.2 **<a id="section-32">AI Model</a>**

**<a id="section-321">Base Model</a>**

จากปัญหาของผมที่ต้องการสร้างโมเดลสำหรับจำแนกรูปภาพเลยใช้ คำสั่งสร้างโมเดลฐาน (base model) ที่มีความสามารถในการดึง feature หรือลักษณะเฉพาะของภาพออกมาได้ดี เช่น สีของเมล็ดกาแฟที่แตกต่างกัน โดยใช้โมเดล Xception ซึ่งเป็นหนึ่งในโมเดล Convolutional Neural Network (CNN) ที่มีประสิทธิภาพสูงในการแยกรูปภาพ และยังปรับใช้ได้ง่าย ซึ่งมีการใช้ฟีเจอร์ที่เรียนรู้จากข้อมูลขนาดใหญ่ เช่น ImageNet การตั้งค่า pooling='max' ช่วยให้โมเดลทำการดึงฟีเจอร์ที่สำคัญออกมาจากภาพได้ดีขึ้น โดยเลือกค่าที่สูงที่สุดจากฟีเจอร์แมพ ซึ่งจะช่วยให้ข้อมูลมีความกระชับและรักษาข้อมูลที่สำคัญ

```python
# สร้าง base model โดยใช้ Xception
base_model = tf.keras.applications.Xception(include_top=False, weights="imagenet",
                                             input_shape=img_shape, pooling='max')
```

**<a id="section-322">Create Model</a>**

ใช้โมเดลแบบ Subclass ใน Keras เพราะผมต้องการให้โมเดลมีความยืดหยุ่นและสามารถปรับให้เข้ากับ Hyperparameters โดยตรง ไม่ต้องผ่าน Sequential, Functional API

```python
class MyModel(tf.keras.Model):
    def __init__(self, base_model, dropout_rate_1, dropout_rate_2, dense_units, num_classes):
        super(MyModel, self).__init__()
        # กำหนดโมเดลพื้นฐาน
        self.base_model = base_model
        # กำหนดเลเยอร์ต่างๆ
        self.flatten = Flatten()
        self.dropout1 = Dropout(rate=dropout_rate_1)
        self.dense = Dense(dense_units, activation='relu')
        self.dropout2 = Dropout(rate=dropout_rate_2)
        self.output_layer = Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.base_model(inputs)
        x = self.flatten(x)
        x = self.dropout1(x)
        x = self.dense(x)
        x = self.dropout2(x)
        return self.output_layer(x)

# การใช้งาน MyModel
model1 = MyModel(base_model=base_model,
                dropout_rate_1=dropout_rate_1,
                dropout_rate_2=dropout_rate_2,
                dense_units=dense_units,
                num_classes=num_classes)
```

**<a id="section-323">Compile Model</a>**

```python
model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

**<a id="section-324">Summary</a>**

```python
model1.build((None, *img_shape))
model1.summary()
```

- อัตราส่วนข้อมูล 70% ใช้ train 30% ใช้ test(ดูหัวข้อด้านบน[Train-Test Split](#section-231))
